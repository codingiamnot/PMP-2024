```python
import pymc as pm
import numpy as np
import arviz as az

obs = [56, 60, 58, 55, 57, 59, 61, 56, 58, 60]

with pm.Model() as model:

  media = pm.Normal("media", mu=np.mean(obs), sigma=100)
  sd = pm.HalfNormal("sd", sigma=10)

  obs = pm.Normal("obs", mu=media, sigma=sd, observed=obs)

  sample = pm.sample(2000, tune=1000, target_accept=0.95, random_seed=1720357)

summary = az.summary(sample, var_names=["media", "sd"])
summary

az.hdi(sample, var_names=["media", "sd"], hdi_prob=0.95)
```
